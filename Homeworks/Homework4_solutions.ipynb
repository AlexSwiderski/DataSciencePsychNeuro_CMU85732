(2) Three methods for evaluating a regression model’s performance are R2, RSE, and the F Test. Show the functional form of each method and describe conceptually what it is telling you about your model.

* **R-square:** $r^2=(TSS-RSS)/TSS=1-(RSS/TSS)=1-\frac{(\sum_{i=1}^{n}(y_i-\hat{y}_i)^2))}{(\sum{i=1}^{n}(y_i-\bar{y}_i)^2)}$. The $r^2$ tells you the percentage of the variance in the response variable y that can be explained by the model.

* **Residual standard error:** $RSE =\frac{(y_i-\hat{y}_i)^2}{(n-2)}=\frac{RSS}{(n-2)}$. The RSE provides a measure of confidence of the model fit. Smaller values of RSE indicate greater confidence that your model explains the data.

* **F Test:** $ F= \frac{\frac{(TSS-RSS)}{p}}{\frac{RSS}{(n-p-1)}} = \frac{\frac{1}{p}(\sum_{i=1}^{n}(y_i-\bar{y}_i)^2 - \sum_{i=1}^{n}(y_i-\hat{y}_i)^2)}{\frac{\sum_{i=1}^{n}(y_i-\hat{y}_i)^2}{n-p-1}}$. If the variance not explained by your model (TSS) is greater than the variance explained your model (RSS), then the ratio is high and F is significant

(4) A linear mixed effect model has the form $Y= X\beta + Z\upsilon + \epsilon$. 

<br>
(a) What are the assumptions about Z and  that make this different from using a regular control variable in a simple linear regression model?

<br>
**Answer:** For linear mixed effects models we assume that Z is a categorical variable whose values are arbitrary (e.g., batch ID, subject ID). We also assume that $\upsilon$ is random as well (i.e., $\upsilon \sim N(0,\sigma)$ ). When using a control variable in simple linear regression, you assume that the regression coefficient for that variable is not random but meaningful in its relationship to the response variable. You also assume in simple regression that the value of the regressor itself maps to a meaningful relationship in the data (e.g., gender, age, education). 

<br>
<br>
(b) What is the objective function for this model and how is it different than the objective function for the ordinary least squares model?

** Answer: ** 

Linear Mixed Effects: $min{||y-X\beta-Z\Lambda_{\theta}\upsilon||^2+||\upsilon||^2}$

Linear Regression: $min{||y-X\beta||^2}$ 

The linear mixed effects objective function includes the covariance matrix of the random effect term itself ($\Lambda_{\theta}$) and adds an additional penalty for low values of $\upsilon$ (i.e., regularizes the fits on the random effects). Whereas the linear regression objective function only attempts to minimize the mean squared error.

**Answer**  It allows for you to account for predictor variables that are not of primary interest and result from random selection, but may still impact the response variable itself.

```r
# (b)
lm.fit = lm(incidence~size)
summary(lm.fit)
par(mfrow=c(2,2))
plot(lm.fit)

# Herd size has a significant impact on incidence of CBPP. As
# the herd increases, animals are more likely to become infected.
# The p-value on the t-statistic for the regression coefficient is 
# less than 0.05.
```

```r

# (c)
lme.fit = lmer(incidence ~ size + (size | herd))
summary(lme.fit)

# The effect of herd size gets stronger when you include 
# the random effect of herd ID. The effect of size in the
# simple regression model is 0.138 and in the mixed effect
# model it is 0.196.

# (d)
AIC(lm.fit, lme.fit)

# The AIC is 10 points lower for the mixed effects model than
# the simple linear regression model. This indicates that it is 
# a better fit despite having more parameters in the model.
```
